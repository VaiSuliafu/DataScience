```python
# imports matplotlib with the usual abbreviation
import matplotlib as mpl

# the version of matplotlib we want to use
mpl.__version__

# imports the main plotitng (sub) package with the usual abbreviation plt
import matplotlib.pyplot as plt

# set plotitng style to seaborn
plt.style.use('seaborn')

#sets the font to be serif in all plots
mpl.rcParams['font.family'] = 'serif'

%matplotlib inline
```

```python
# imports numpy with the usual abbreviation
import numpy as np

# Fixes the seed for the random number generator for reproducibility
np.random.seed(1000)

# Draws the random y-values
y = np.random.standard_normal(20)

# Fixes the x-value integers
x = np.arange(len(y))

# Calls the plt.plot() function with the x and y objects
plt.plot(x,y)
```

```python
# Now plotting the cumulative sums
plt.plot(y.cumsum())
```

```python
# Now re-plotting with some custom configuration
plt.plot(y.cumsum())

# Turns off the grid
plt.grid(False)

# Leads to equal scaling for the 2 axes
plt.axis('Equal')
```

```python
# Now a plot with custom axis limits
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1,
         np.max(y.cumsum()) + 1)
```

```python
# Now a plot with labels for improved readability
plt.figure(figsize = (10, 6))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title("A Simple Plot")
```

```python
# Now generating a sample two-dimensional nd array object with standard normally disbributed pseudo-random numbers
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
# Now passing the 2-d array to plot.plot()
plt.figure(figsize = (10, 6))
plt.plot(y, lw = 1.5)
plt.plot(y, 'ro')
plt.xlabel("Index")
plt.ylabel("Values")
plt.title("A Simple Plot")
```

```python
# Now to make the sample plot with some added labels for clarity
plt.figure(figsize = (10,6))
plt.plot(y[:, 0], 'b', lw = 1.5, label = "1st")
plt.plot(y[:, 1], 'r', lw = 1.5, label = "2nd")
plt.plot(y, 'ko')
plt.legend(loc=0) # places the legend in the 'best' location
plt.xlabel('Index')
plt.ylabel('Value')
plt.title("A Simple Plot")
```

```python
# Now to do a 2D plot with dual y-axes

# Defining the figure and axis objects
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw = 1.5, label = "1st")
plt.plot(y[:, 0], 'ko')
plt.legend(loc = 8)
plt.xlabel('Index')
plt.ylabel('Value 1st')
plt.title('A Simple Plot')

# Creates a second axis object that shares the x-axis
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'r', lw = 1.5, label = "2nd")
plt.plot(y[:, 1], 'ko')
plt.legend(loc = 0)
plt.ylabel("Value 2nd")
```

```python
# Now to display using two separate subplots
plt.figure(figsize = (10,6))
plt.subplot(2,1,1) # defines the upper subplot
plt.plot(y[:, 0], 'k', lw = 1.5, label = "1st")
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 0)
plt.ylabel('Value')
plt.title('A Simple Plot')
plt.subplot(2,1,2) # defines the lower subplot
plt.plot(y[:, 1], 'k', lw = 1.5, label = "2nd")
plt.plot(y[:, 1], 'ro')
plt.legend(loc = 0)
plt.xlabel('Index')
plt.ylabel('Value')
```

```python
# Using subplots to combine two different plot types
plt.figure(figsize = (10,6))
plt.subplot(1,2,1)
plt.plot(y[:, 0], lw = 1.5, label = "1st")
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title("1st Data Set")

plt.subplot(1,2,2)
plt.bar(np.arange(len(y)), y[:,1], width = 0.5, color = 'g', label = "2nd") # creates the bar subplot
plt.legend(loc = 0)
plt.xlabel('Index')
plt.title('2nd Data Set')
```
